```python
with open("day11.txt") as f:
    input = f.read().splitlines()
```

```python
fulL_input = []
for row in input: ## THING TO CHANGE INPUT IN
    fulL_input.append([*row])
```

```python

def add_expansion(list_of_lists):
    added_rows = []
    for row in list_of_lists:
        if all(x == '.' for x in row):
            added_rows.append(row)
        added_rows.append(row)
    return added_rows

def note_expanded_rows(list_of_lists):
    clear_rows = []
    for i,row in enumerate(list_of_lists):
        if all(x == '.' for x in row):
            clear_rows.append(i)
    return clear_rows

clear_rows = note_expanded_rows(fulL_input)
clear_rows
```

```python
flipped = np.array(fulL_input).T.tolist()
```

```python
clear_columns = note_expanded_rows(flipped)
clear_columns
```

```python
galaxies = []
for row_number,row in enumerate(fulL_input):
    for column_number,point in enumerate(row):
        if point =='#':
            galaxies.append((row_number,column_number))
galaxies
```

```python
total = 0
for i, galaxy in enumerate(galaxies):
    for j in range(i+1,len(galaxies)):
        num_clear_rows = len([r for r in clear_rows if (r>min(galaxy[0],galaxies[j][0]) and r<max(galaxy[0],galaxies[j][0]))])
        num_clear_columns = len([c for c in clear_columns if (c>min(galaxy[1],galaxies[j][1]) and c<max(galaxy[1],galaxies[j][1]))])
        extra_distance_rows = num_clear_rows*(1000000-1)
        extra_distance_columns = num_clear_columns*(1000000-1)
        total+= abs(galaxy[0]-galaxies[j][0])+abs(galaxy[1]-galaxies[j][1]) + extra_distance_columns + extra_distance_rows
```
